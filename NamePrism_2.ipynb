```python
import ast
import requests
from urllib.parse import quote
import time

nameprism_list = []

for i in np.array(gender_data.index):
    url = 'https://www.name-prism.com/api_token/eth/json/56865f8ac15dc407/' + quote(data.loc[i, 'name'])
    res = requests.get(url)
    ret_dict = ast.literal_eval(res.text)
    ret_dict['index'] = i
    nameprism_list.append(ret_dict)
    time.sleep(2)
```
